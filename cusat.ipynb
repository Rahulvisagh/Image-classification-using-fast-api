```python
import os
from fastai.vision.all import *
from fastai.vision import *
import pandas as pd
```

```python
Dataset_path='/content/drive/MyDrive/images/im'
```

```python
class_names = []
for folder_name in os.listdir(Dataset_path):
  class_names.append(folder_name)
```

```python
class_names
```

```python
os.listdir(Dataset_path)
```

```python
#Load the data
dls = ImageDataLoaders.from_folder(
    path=Dataset_path,
    item_tfms=Resize(224),
    bs=16,
    batch_tfms=[Normalize.from_stats(*imagenet_stats), RandTransform()],
    valid_pct=0.2
)
```

```python
dls.show_batch()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#create model
model=vision_learner(
    dls=dls,
    arch=models.resnet50,
    metrics=[accuracy,error_rate]
)
```

```python
model.fit(5)
```

```python
interp = ClassificationInterpretation.from_learner(model)
interp.plot_confusion_matrix()
```

```python
model.show_results()
```

```python
#save model
model.export('home.pkl')
```

```python
#load the model
new_model = load_learner('/content/drive/MyDrive/images/im/home.pkl')
```

```python
from fastai.vision.all import*
test_path = Path("/content/drive/MyDrive/test")
test_items = get_image_files(test_path)

test_dl = new_model.dls.test_dl(test_items,with_labels=True)

preds,target=new_model.get_preds(dl=test.dl)

accuracy = new_model.validate(dl=test_dl)
accuracy[0]
```
